## Использование панелей при визуализации

```python
ggplot(data=mpg) +\
geom_point(mapping=aes(x="displ", y="hwy")) +\
facet_wrap("class", nrow=2)
```

```python
ggplot(data=mpg) +\
geom_point(mapping=aes(x="displ", y="hwy")) +\
facet_grid("drv ~ cyl")
```

### Что если использовать категорию с непрерывными значениями (например, hwy) для разделения на панели?

```python
ggplot(data=mpg) + \
    geom_point(mapping=aes(x="displ", y="class")) + \
    facet_wrap("hwy")
```

### Можно сделать только вертикальные или только горизонтальные панели, если использовать конструкцию типа (". ~ <категория>")

```python
ggplot(data=mpg) + \
    geom_point(mapping=aes(x="displ", y="hwy")) + \
    facet_grid(". ~ cyl")
```

```python
ggplot(data=mpg) + \
    geom_point(mapping=aes(x="displ", y="hwy")) + \
    facet_grid("drv ~ .")
```

### Что лучше использовать: цвета или панели? А для большого набора?

Для меленького набора презентабельней использовать цвета.
Для большого лучше панели, также они хороши, если хочется посмотреть на большее число зависимостей, к примеру график зависимости расхода топлива в городе и на трассе в зависимости от года и объема двигателя лучше рассмотреть на панели.

```python
ggplot(data=mpg) + \
    geom_point(mapping=aes(x="hwy", y="cty", color="displ")) + \
    facet_wrap("year")
```

### Что делает geom_col()? Чем он отличается от geom_bar()?

```python
diamonds
```

```python
ggplot(data=diamonds) +\
geom_col(mapping=aes(x="cut", y="carat"))
```

geom_col() составляет высоту столбцов в зависимости от значений y

geom_bar() составляет высоту в зависимости от встречаемости значений в x и не требует y

### Что произойдёт, если 
'ggplot(data=diamonds) +\
'geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
или
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", fill="color", y="..prop.."))
### запустить без "group=1"? Почему?

```python
ggplot(data=diamonds) \
    + geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
```

```python
ggplot(data=diamonds) \
    + geom_bar(mapping=aes(x="cut", fill="color", y="..prop.."))
```

При неиспользовании group=1 все столбцы выравниваются, так как не указано по какой переменной мы считаем частоту для задачи высоты столбцов

Мешает значение, которое записано в y

### Зачем тут coord_fixed()? Что делает geom_abline()?

geom_abline() - рисует линию, которая является неким правилом, не зависит от данных

coord_fixed() - фиксирует масштам системы координат

```python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) +\
geom_point() +\
geom_abline() +\
coord_fixed()
```
